```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.impute import SimpleImputer
```

```python
# Load the datasets
train_data = pd.read_csv('/content/drive/MyDrive/CodSoft/task-2/fraudTrain.csv')
test_data = pd.read_csv('/content/drive/MyDrive/CodSoft/task-2/fraudTest.csv')
```

```python
# Select relevant features for the model
features = ['cc_num', 'merchant', 'category', 'amt', 'gender', 'zip', 'lat', 'long', 'city_pop', 'unix_time', 'merch_lat', 'merch_long']
```

```python
# Separate features and target variable from training data
X_train = train_data[features]
y_train = train_data['is_fraud']
```

```python
# Separate features and target variable from test data
X_test = test_data[features]
y_test = test_data['is_fraud']
```

```python
# Preprocessing steps for numeric and categorical data
numeric_features = ['amt', 'lat', 'long', 'city_pop', 'unix_time', 'merch_lat', 'merch_long']
categorical_features = ['merchant', 'category', 'gender']
```

```python
# Create a column transformer with a pipeline for both numeric and categorical features
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
])
```

```python
# Create a pipeline that includes preprocessing and the logistic regression model
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(solver='liblinear'))
])
```

```python
# Train the model
pipeline.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = pipeline.predict(X_test)
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)
```

```python
print(f'Accuracy: {accuracy}')
print('Confusion Matrix:')
print(conf_matrix)
print('Classification Report:')
print(class_report)
```
